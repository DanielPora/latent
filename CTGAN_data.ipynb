```python
from ctgan import CTGAN
from ctgan import load_demo
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./clean_F.csv")
data = data.drop(columns=['Unnamed: 0','workerID'], axis=1)
data
```

```python
data = data.dropna(axis=1, how='any')
data
```

```python
discrete = ['perspective',
           'targetObj',
           'targetPos',
           'respObjPos',
           'gender',
            'responderType'
           ]

```

```python
cont_data = data.drop(columns=['perspective',
           'targetObj',
           'targetPos',
           'respObjPos',
           'gender',
            'responderType'
           ], axis=1)
cont_data
```

```python
ctgan = CTGAN(epochs=1)
ctgan.fit(cont_data)
synth_cont = ctgan.sample(10)
```

```python
synth_cont
```

```python
ctgan = CTGAN(epochs=1000)
ctgan.fit(data, discrete)
```

```python
synth = ctgan.sample(100)
synth.to_csv("synth1000.csv")
```
